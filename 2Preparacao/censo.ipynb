```python
# importando bibliotecas
import pandas as pd
import numpy as np
```

```python
df_2019 = pd.read_csv('../../Dados/Censo/populacao_raca_2019.csv')
df_2020 = pd.read_csv('../../Dados/Censo/populacao_raca_2020.csv')
df_2021 = pd.read_csv('../../Dados/Censo/populacao_raca_2021.csv')
```

```python
df_populacao = pd.concat([df_2019,df_2020,df_2021],ignore_index=True)
df_populacao
```

```python
# Como os valores são para cada 1000 pessoas, multiplicamos os valores populacionais por 1000
df_populacao['total_pessoas'] = df_populacao['total_pessoas'] * 1000


# Criamos novas colunas dos dados que estão em formato percentual para numérico
df_populacao['pop_branca'] = round(((df_populacao['perc_branca']/100)) * df_populacao['total_pessoas'],0)
df_populacao['pop_preta'] = round(((df_populacao['perc_preta']/100)) * df_populacao['total_pessoas'],0)
df_populacao['pop_parda'] = round(((df_populacao['perc_parda']/100)) * df_populacao['total_pessoas'],0)
df_populacao['pop_preta_ou_parda'] = round(((df_populacao['perc_preta_ou_parda']/100)) * df_populacao['total_pessoas'],0)

df_populacao
```

```python
#Incluindo uma coluna com os nomes dos estados em extenso

df_populacao['estado'] = df_populacao['uf']
mapeamento = {
    "AC":"Acre",
    "AL":"Alagoas",
    "AP":"Amapá",
    "AM":"Amazonas",
    "BA":"Bahia",
    "CE":"Ceará",
    "DF":"Distrito Federal",
    "ES":"Espírito Santo",
    "GO":"Goiás",
    "MA":"Maranhão",
    "MT":"Mato Grosso",
    "MS":"Mato Grosso do Sul",
    "MG":"Minas Gerais",
    "PA":"Pará",
    "PB":"Paraíba",
    "PR":"Paraná",
    "PE":"Pernambuco",
    "PI":"Piauí",
    "RJ":"Rio de Janeiro",
    "RN":"Rio Grande do Norte",
    "RS":"Rio Grande do Sul",
    "RO":"Rondônia",
    "RR":"Roraima",
    "SC":"Santa Catarina",
    "SP":"São Paulo",
    "SE":"Sergipe",
    "TO":"Tocantins"
    
}

df_populacao['estado'] = df_populacao['estado'].replace(mapeamento)
df_populacao
```

```python
# Melt the DataFrame to transform it into the desired format
melted_df = pd.melt(df_populacao, id_vars=['ano', 'uf','estado'], value_vars=['pop_branca', 'pop_preta', 'pop_parda'],
                    var_name='racacor', value_name='TOTAL')

melted_df
```

```python
# Transformação do data frame em um novo agrupado pelos atributos que desejamos
melted_df['percentage_of_total'] = (melted_df['TOTAL'] / melted_df.groupby(['ano', 'uf','estado'])['TOTAL'].transform('sum')) * 100
melted_df
```

```python
# Conversão do dataframe em CSV
melted_df.to_csv("../Dados/Dataset_Power_BI/Dados_Censo.csv")
```
