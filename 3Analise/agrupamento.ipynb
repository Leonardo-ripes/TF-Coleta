```python
import pandas as pd
```

```python
# União dos dois CSVs tratados em um novo df unicado

dfpop = pd.read_csv("../Dados/Agrupamentos/Dados_Censo.csv")
dfsus = pd.read_csv("../Dados/Agrupamentos/Dados_PYSUS.csv")
```

```python
#Incluindo uma coluna com os nomes dos estados em extenso
mapeamento = {
    "AC":"Acre",
    "AL":"Alagoas",
    "AP":"Amapá",
    "AM":"Amazonas",
    "BA":"Bahia",
    "CE":"Ceará",
    "DF":"Distrito Federal",
    "ES":"Espírito Santo",
    "GO":"Goiás",
    "MA":"Maranhão",
    "MT":"Mato Grosso",
    "MS":"Mato Grosso do Sul",
    "MG":"Minas Gerais",
    "PA":"Pará",
    "PB":"Paraíba",
    "PR":"Paraná",
    "PE":"Pernambuco",
    "PI":"Piauí",
    "RJ":"RJ",
    "RN":"Rio Grande do Norte",
    "RS":"Rio Grande do Sul",
    "RO":"Rondônia",
    "RR":"Roraima",
    "SC":"Santa Catarina",
    "SP":"SP",
    "SE":"Sergipe",
    "TO":"Tocantins"
}

result_df['UF'] = result_df['UF'].replace(mapeamento)
result_df
```

```python
result_df = pd.merge(dfpop, dfsus, on=['ANO', 'UF','RACACOR'], how='inner')

#Rename columns for clarity
result_df = result_df.rename(columns={
    'TOTAL_PESSOAS_x': 'TOTAL_POPULATION',
    'TOTAL_PESSOAS_y': 'TOTAL_SUICIDALS',
    'CENSO_total/sum_pessoas_racacor': 'PERCENTAGE_TOTAL_POPULATION',
    'sim_total/sum_pessoas_racacor': 'PERCENTAGE_TOTAL_SUICIDALS'
}).drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)

result_df
```

```python
result_df['PERCENTAGE_SUICIDALS'] = result_df['TOTAL_SUICIDALS'] / result_df['TOTAL_POPULATION']
```

```python
df_rs = result_df.where((result_df['UF'] == 'Rio Grande do Sul') & (result_df['ANO'] == 2021)).dropna()
df_rs['PERCENTAGE_SUICIDALS'].sum() * 100
```

```python
# Convertendo o dataframe em csv 
result_df.to_csv("../Dados/Agrupamentos/Dados_Mergados.csv")
```
