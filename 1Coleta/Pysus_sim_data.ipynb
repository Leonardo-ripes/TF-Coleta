```python
from pysus.online_data.SIM import download
import pandas as pd
import numpy as np
from datetime import datetime

estados = [
    "AC",  # Acre
    "AL",  # Alagoas
    "AP",  # Amapá
    "AM",  # Amazonas
    "BA",  # Bahia
    "CE",  # Ceará
    "DF",  # Distrito Federal
    "ES",  # Espírito Santo
    "GO",  # Goiás
    "MA",  # Maranhão
    "MT",  # Mato Grosso
    "MS",  # Mato Grosso do Sul
    "MG",  # Minas Gerais
    "PA",  # Pará
    "PB",  # Paraíba
    "PR",  # Paraná
    "PE",  # Pernambuco
    "PI",  # Piauí
    "RJ",  # Rio de Janeiro
    "RN",  # Rio Grande do Norte
    "RS",  # Rio Grande do Sul
    "RO",  # Rondônia
    "RR",  # Roraima
    "SC",  # Santa Catarina
    "SP",  # São Paulo
    "SE",  # Sergipe
    "TO",  # Tocantins
]
anos = [2019,2020,2021]
COLUMNS_TO_SELECT = ['RACACOR','SEXO','DTOBITO', 'DTNASC', 'UF', 'IDADE_ANOS','CIRCOBITO']
all_states_df = None

for y in anos:
  cleaned_dataframes = []
  for uf in estados:
    try:
      print(f'starting execution to uf: {str(uf)} and year: {str(y)}')
      started_at = datetime.now()
      file_path = download(uf, y)
      dataframe = pd.read_parquet(file_path)
      dataframe = dataframe[dataframe['CIRCOBITO'] == '2']
      dataframe['UF'] = uf
      dataframe['YEAR'] = y
      dataframe['DTNASC'] = dataframe['DTNASC'].str.strip().replace('', np.nan)
      dataframe['DTOBITO'] = dataframe['DTOBITO'].str.strip().replace('', np.nan)
      dataframe['DTOBITO'] = pd.to_datetime(dataframe['DTOBITO'], format='%d%m%Y')
      dataframe['DTNASC'] = pd.to_datetime(dataframe['DTNASC'], format='%d%m%Y')
      dataframe['IDADE_ANOS'] = dataframe['DTOBITO'].dt.year - dataframe['DTNASC'].dt.year
      dataframe = dataframe[COLUMNS_TO_SELECT]
      dataframe = dataframe.dropna().reset_index(drop=True)
      cleaned_dataframes.append(dataframe)
      ended_at = datetime.now()
      print(f'done! time: {ended_at - started_at} and total rows: {dataframe.shape[0]}')
    except Exception as e:
      # Handle other exceptions if they occur.
      print(f"An unexpected error occurred: {str(e)}")
  all_states_df = pd.concat(cleaned_dataframes, axis=0).reset_index(drop=True)
  print(f'total_rows: {str(all_states_df.shape[0])}')
  all_states_df.to_csv(f'sim_data_cleaned_{str(y)}.csv')
```
